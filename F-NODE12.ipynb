# 막대그래프 그리기

```python
import matplotlib.pyplot as plt
%matplotlib inline  # graph 바로출력됨
```

```python
#그래프 데이터
subject = ['English', 'Math', 'Korean', 'Science', 'Computer']
points = [40, 90, 50, 60, 100]
```

축 그리기

```python
# 축 그리기
fig = plt.figure() # 그림객체(figure) 
ax1 = fig.add_subplot(1,1,1) # 그 안에 하위그래프(subplot)추가
```

```python
flg = plt.figure(figsize=(5,2)) #figsize 그래프 크기 정함
ax1 = flg.add_subplot(1,1,1) #축 설정(그래프 갯수?)
```

```python
flg = plt.figure()
ax1 = flg.add_subplot(2,2,1) # 2행 2열의 1번째 위치
ax2 = flg.add_subplot(2,2,2)
ax4 = flg.add_subplot(2,2,4)
```

그래프 그리기

```python
#그래프데이터 >>위에

#축그리기 >> 
fig = plt.figure() 
ax1 = fig.add_subplot(1,1,1) 

#그래프 그리기 (x,y순)
ax1.bar(subject, points)

#그래픽 요소추가(label, title)
plt.xlabel('Subject')
plt.ylabel('Points')
plt.title("Yuna's Test result")
```

# 선 그래프 그리기

```python
from datetime import datetime
import pandas as pd
import os
```

```python
os.environ['HOME']
```

```python
# os.environ["HOME"] = os.getenv("HOME")
```

```python
#그래프 데이터
csv_path = os.getenv("HOME") + "/aiffel/data_represent/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['Close']
```

```python
data.head()
```

```python
# 축 그리기 및 좌표축 설정
flg = plt.figure(figsize=(10,4))
ax = flg.add_subplot (1,1,1)
price.plot(ax=ax, style='black')
plt.ylim([1600,2200])
plt.xlim(['2019-05-01','2020-03-01'])

# 주석달기 annotate()
important_data = [(datetime(2019,6,3),'Low Price'),(datetime(2020,2,19),'Peak Price')]
for d, label in important_data:
    print(d)
    print(label)
    ax.annotate(label, xy=(d, price.asof(d)+10),
               xytext=(d,price.asof(d)+100),
               arrowprops=dict(facecolor='red'))
    
# 그리드, 타이틀 달기
plt.grid() # 격자눈금 추가
ax.set_title('StockPrice')

# 보여주기
plt.show()
```

# plt.plot 사용

```python
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x),'o')
plt.plot(x, np.cos(x), '--', color='black')
plt.show()
```

```python
# plt.subplot 가능
x = np.linspace(0, 10, 100)

plt.subplot(2,1,1)
plt.plot(x, np.sin(x),'orange','o')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), 'orange') 
plt.show()
```

```python
x = np.linspace(0, 10, 100) 

plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='dashed') 
plt.plot(x, x + 2, linestyle='dashdot') 
plt.plot(x, x + 3, linestyle='dotted')
plt.plot(x, x + 0, '-g') # solid green 
plt.plot(x, x + 1, '--c') # dashed cyan 
plt.plot(x, x + 2, '-.k') # dashdot black 
plt.plot(x, x + 3, ':r'); # dotted red
plt.plot(x, x + 4, linestyle='-') # solid 
plt.plot(x, x + 5, linestyle='--') # dashed 
plt.plot(x, x + 6, linestyle='-.') # dashdot 
plt.plot(x, x + 7, linestyle=':'); # dotted
```
