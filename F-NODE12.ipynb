그래프 그리기

# 선 그래프 그리기

```python
from datetime import datetime
import pandas as pd
import os
```

```python
os.environ['HOME']
```

```python
# os.environ["HOME"] = os.getenv("HOME")
```

```python
#그래프 데이터
csv_path = os.getenv("HOME") + "/aiffel/data_represent/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['Close']
```

```python
data.head()
```

```python
# 축 그리기 및 좌표축 설정
flg = plt.figure(figsize=(10,4))
ax = flg.add_subplot (1,1,1)
price.plot(ax=ax, style='black')
plt.ylim([1600,2200])
plt.xlim(['2019-05-01','2020-03-01'])

# 주석달기 annotate()
important_data = [(datetime(2019,6,3),'Low Price'),(datetime(2020,2,19),'Peak Price')]
for d, label in important_data:
    print(d)
    print(label)
    ax.annotate(label, xy=(d, price.asof(d)+10),
               xytext=(d,price.asof(d)+100),
               arrowprops=dict(facecolor='red'))
    
# 그리드, 타이틀 달기
plt.grid() # 격자눈금 추가
ax.set_title('StockPrice')

# 보여주기
plt.show()
```

# plt.plot 사용

```python
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x),'o')
plt.plot(x, np.cos(x), '--', color='black')
plt.show()
```

```python
# plt.subplot 가능
x = np.linspace(0, 10, 100)

plt.subplot(2,1,1)
plt.plot(x, np.sin(x),'orange','o')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), 'orange') 
plt.show()
```

```python
x = np.linspace(0, 10, 100) 

plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='dashed') 
plt.plot(x, x + 2, linestyle='dashdot') 
plt.plot(x, x + 3, linestyle='dotted')
plt.plot(x, x + 0, '-g') # solid green 
plt.plot(x, x + 1, '--c') # dashed cyan 
plt.plot(x, x + 2, '-.k') # dashdot black 
plt.plot(x, x + 3, ':r'); # dotted red
plt.plot(x, x + 4, linestyle='-') # solid 
plt.plot(x, x + 5, linestyle='--') # dashed 
plt.plot(x, x + 6, linestyle='-.') # dashdot 
plt.plot(x, x + 7, linestyle=':'); # dotted
```

### 데이터 살펴보기(EDA)

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
```

```python
# category :  범주형 데이터 , + size(테이블인원)
cate = ['sex', 'time', 'smoker', 'day', 'size']
```

### 범주형 데이터: 주로 막대 그래프를 사용하여 수치를 요약. 
* 일반적으로 가로, 세로, 누적, 그룹화된 막대 그래프를 사용.

```python
# tip을 성별에 대한 평균으로
grouped = df['tip'].groupby(df['sex'])
```

```python
grouped.mean() # 성별에 따른 팁의 평균
```

```python
grouped.size() # 성별에 따른 데이터 량(팁 횟수)
```

```python
dict(grouped.mean())
```

```python
dict(grouped.mean()).keys()
```

```python
import numpy as np
sex = dict(grouped.mean())
x = list(sex.keys())
y = list(sex.values())

print(x)
print(y)

plt.bar(x,y)
plt.ylabel(['tip[$]'])
plt.title('Tip by Sex')
plt.legend()
```

```python
sns.barplot(data=df, x='sex',y='tip')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='sex', y='tip')
plt.ylim(0, 4)
plt.title('Tip by sex')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='day', y='tip')
plt.ylim(0, 4)
plt.title('Tip by day')
```

```python
fig = plt.figure(figsize=(13,10)) 
ax1 = fig.add_subplot(2,2,1)
sns.barplot(data=df, x='day', y='tip',palette="ch:.25") #palette = 색깔
ax2 = fig.add_subplot(2,2,2)
sns.barplot(data=df, x='sex', y='tip')
ax4 = fig.add_subplot(2,2,4)
sns.violinplot(data=df, x='sex', y='tip')
ax3 = fig.add_subplot(2,2,3)
sns.violinplot(data=df, x='day', y='tip',palette="ch:.25")
```

```python
sns.catplot(x="day", y="tip", jitter=False, data=tips) #jitter : 퍼짐정도
```
