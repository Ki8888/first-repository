```python
#그래프 데이터
csv_path = os.getenv("HOME") + "/aiffel/data_represent/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['Close']
```

```python
data.head()
```

```python
# 축 그리기 및 좌표축 설정
flg = plt.figure(figsize=(10,4))
ax = flg.add_subplot (1,1,1)
price.plot(ax=ax, style='black')
plt.ylim([1600,2200])
plt.xlim(['2019-05-01','2020-03-01'])

# 주석달기 annotate()
important_data = [(datetime(2019,6,3),'Low Price'),(datetime(2020,2,19),'Peak Price')]
for d, label in important_data:
    print(d)
    print(label)
    ax.annotate(label, xy=(d, price.asof(d)+10),
               xytext=(d,price.asof(d)+100),
               arrowprops=dict(facecolor='red'))
    
# 그리드, 타이틀 달기
plt.grid() # 격자눈금 추가
ax.set_title('StockPrice')

# 보여주기
plt.show()
```

### 범주형 데이터: 주로 막대 그래프를 사용하여 수치를 요약. 
* 일반적으로 가로, 세로, 누적, 그룹화된 막대 그래프를 사용.

```python
# tip을 성별에 대한 평균으로
grouped = df['tip'].groupby(df['sex'])
```

```python
grouped.mean() # 성별에 따른 팁의 평균
```

```python
grouped.size() # 성별에 따른 데이터 량(팁 횟수)
```

```python
dict(grouped.mean())
```

```python
dict(grouped.mean()).keys()
```

```python
import numpy as np
sex = dict(grouped.mean())
x = list(sex.keys())
y = list(sex.values())

print(x)
print(y)

plt.bar(x,y)
plt.ylabel(['tip[$]'])
plt.title('Tip by Sex')
plt.legend()
```

```python
sns.barplot(data=df, x='sex',y='tip')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='sex', y='tip')
plt.ylim(0, 4)
plt.title('Tip by sex')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='day', y='tip')
plt.ylim(0, 4)
plt.title('Tip by day')
```

* 히스토그램

```python
# 그래프 데이터
mu1, mu2, sigma = 100,130,15
x1=mu1+sigma*np.random.randn(10000)
x2=mu2+sigma*np.random.randn(10000)

# 축 그리기
flg = plt.figure()
ax1 = flg.add_subplot(1,1,1)

# 그래프 그리기
patches = ax1.hist(x1, bins=50, density=False) #density>>비율변경, 총량대비데이터비율로
patches = ax1.hist(x2, bins=50, density=False, alpha=0.5)

ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

# 라벨, 타이틀 달기
plt.xlabel('Bins')
plt.ylabel('Numbber of Values in Bin')
ax1.set_title('Two Frequency Distributions')

# 보여주기
plt.show()
```

# HeatMap

* 방대한 양의 데이터와 현상을 수치에 따른 색상으로 나타내는 것

```python
# heatmap 사용위해 pivot 전환 필요
pivot = flights.pivot(index='year', columns='month', values='passengers')
pivot
```

```python
sns.heatmap(pivot)
```

```python
sns.heatmap(pivot, linewidth=.2, annot=True, fmt='d')
```

```python
sns.heatmap(pivot, cmap="YlGnBu")
```
