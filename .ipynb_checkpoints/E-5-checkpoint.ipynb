## ÏñºÍµ¥ Í≤ÄÏ∂ú

### Object dectectionÏù¥Ïö©ÌïòÏó¨ ÏñºÍµ¥Í≤ÄÏ∂ú

* Îã®ÏùºÌîΩÏÖÄÏù¥ ÏïÑÎãå ÏùºÏ†ïÏñëÏùò ÌîΩÏÖÄ(16X16)ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÏÉâÏÉÅÏùò Î≥ÄÌôîÏú®(Î∞ùÏùÄÎ∞©Ìñ•ÏóêÏÑú Ïñ¥ÎëêÏö¥Î∞©Ìñ•)ÏúºÎ°ú gradientÏ∏°Ï†ï

```python
# hog dectectorÏÑ†Ïñ∏(ÏñºÍµ¥Í≤ÄÏ∂ú ÌîÑÎ°úÍ∑∏Îû®)
import dlib
detector_hog = dlib.get_frontal_face_detector() #detectorÏÑ†Ïñ∏
print('Not enough')
```

```python
# ÏñºÍµ¥Ïùò bounding box Ï∂îÏ∂ú
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)
#image pyramids Í∞ôÏùÄÏù¥ÎØ∏ÏßÄÎ•º Îã§Î•∏ÏÇ¨Ïù¥Ï¶àÎ°ú ÏûëÏóÖ, upsamplingÏúºÎ°ú ÌÅ¨Í∏∞Î•º ÌÇ§ÏõÄ (ÏûëÏùÄ ÏñºÍµ¥Ïù¥ Ïª§ÏßÄÎØÄÎ°ú Ï†ïÌôïÌïúÍ≤ÄÏ∂úÍ∞ÄÎä•)
print("Now loading")
```

```python
# ÏñºÍµ¥ Ï∂úÎ†•
print(dlib_rects) # Ï∞æÏùÄ ÏñºÍµ¥ÏòÅÏó≠ Ï¢åÌëú (point4Í≥≥Ïù¥ ÏûàÏñ¥Ïïº squareÏÉùÏÑ±Í∞ÄÎä•?)
for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    cv2.rectangle(img_show,(l,t),(r,b),(0,255,0),2, lineType=cv2.LINE_AA)
    
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

## ÏñºÍµ¥ ÎûúÎìúÎßàÌÅ¨ (face landmark)

 Í∞ùÏ≤¥ ÎÇ¥Î∂ÄÏùò Ï†êÏùÑ Ï∞æÎäîÍ∏∞Ïà† (Object keypoint estimation)
- top-down : bounding boxÎ•º Ï∞æÍ≥† ÎÇ¥Î∂ÄÏùò key point ÏòàÏ∏°
- bottom-up : Ïù¥ÎØ∏ÏßÄ Ï†ÑÏ≤¥Ïùò keypointÎ•º Î®ºÏ†Ä Ï∞æÍ≥† pointÍ¥ÄÍ≥ÑÎ•º Ïù¥Ïö©Ìï¥ÏÑú Íµ∞ÏßëÌôîÌõÑ Î∞ïÏä§ÏÉùÏÑ±

* DlibÎäî ibug300-W datesetÏúºÎ°ú ÌïôÏäµÌïú ÏÇ¨Ï†ÑÌïôÏäµÎ™®Îç∏ Ï†úÍ≥µ  
* ibug300-W datasetÏùÄ 68Í∞úÏùò key point ÏÇ¨Ïö©

```python
# load landmark model
import os
model_path = os.getenv("HOME")+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
print("it's OK")
```

* landmark_predictor Îäî RGB Ïù¥ÎØ∏ÏßÄÏôÄ dlib.rectangleÏùÑ ÏûÖÎ†• Î∞õÏïÑ dlib.full_object_detection Î•º Î∞òÌôòÌï©ÎãàÎã§.

```python
list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p:(p.x,p.y), points.parts()))
    list_landmarks.append(list_points)
    
print(len(list_landmarks[0]))
```

* points Îäî dlib.full_object_detection Ïùò Í∞ùÏ≤¥Ïù¥Í∏∞ ÎïåÎ¨∏Ïóê parts() Ìï®ÏàòÎ°ú Í∞úÎ≥Ñ ÏúÑÏπòÏóê Ï†ëÍ∑ºÌï† Ïàò ÏûàÏäµÎãàÎã§. Ï°∞Í∏à Îçî ÏßÅÍ¥ÄÏ†ÅÏù∏ (x, y) ÌòïÌÉúÎ°ú Ï†ëÍ∑ºÌï† Ïàò ÏûàÎèÑÎ°ù Î≥ÄÌôòÌï¥ Ï£ºÏóàÏäµÎãàÎã§. Îî∞ÎùºÏÑú list_points Îäî tuple (x, y) 68Í∞úÎ°ú Ïù¥Î£®Ïñ¥ÏßÑ Î¶¨Ïä§Ìä∏Í∞Ä Îê©ÎãàÎã§. Ïù¥ÎØ∏ÏßÄÏóêÏÑú Ï∞æÏïÑÏßÑ ÏñºÍµ¥ Í∞úÏàòÎßàÎã§ Î∞òÎ≥µÌïòÎ©¥ list_landmarkÏóê 68Í∞úÏùò ÎûúÎìúÎßàÌÅ¨Í∞Ä ÏñºÍµ¥ Í∞úÏàòÎßåÌÅº Ï†ÄÏû•Îê©ÎãàÎã§.(??)

```python
# ÎûúÎìúÎßàÌÅ¨Î•º ÏòÅÏÉÅÏóê Ï∂úÎ†•
for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0,255,255), -1) # yellow
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

## Ïä§Ìã∞Ïª§ Ï†ÅÏö©ÌïòÍ∏∞

* ÎûúÎìúÎßàÌÅ¨Î•º Í∏∞Ï§ÄÏúºÎ°ú ÏΩîÍ∏∞Ï§Ä ÏúÑÏ™ΩÏóê(ÏñºÍµ¥Ï§ëÏïô) Ïä§Ìã∞Ïª§Î•º ÏîåÏö∞Î©¥ Îê†Í±∞Í∞ôÎã§.

```python
# Ï¢åÌëúÌôïÏù∏
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # nose center index : 30
    x = landmark[30][0]
    y = landmark[30][1] - dlib_rect.width()//2
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

```python
# Ïä§Ìã∞Ïª§ Ï†ÅÏö©
import os
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/king.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h)) #ÏúÑÏóêÏÑú Í≥ÑÏÇ∞Ìïú ÌÅ¨Í∏∞Î°ú resize ÎûúÎìúÎßàÌÅ¨ Î≤óÏñ¥ÎÇòÏßÄÏïäÍ≤å..
print (img_sticker.shape)
```

```python
plt.imshow(img_sticker)
```

```python
refined_x = x - w // 2  # left
refined_y = y - h       # top
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

* -49ÎßåÌÅº Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏÇ¨ÏßÑÏùÑ Î≤óÏñ¥ÎÇòÎØÄÎ°ú Í∑∏Í≤ÉÏùÑ ÏûòÎùºÎÇ∏Î∂ÄÎ∂ÑÏùÑ 0ÏúºÎ°ú ÏÇºÏúºÎ†§Î©¥ 49Î•º ÏãúÏûëÏ†êÏúºÎ°ú Ïû°ÏïÑÏïºÌï®

* yÏ∂ï Ï¢åÌëúÍ∞íÏù¥ ÏùåÏàòÎ°ú ÎÇòÏò¥ : ÏõêÎ≥∏Ïù¥ÎØ∏ÏßÄÏùò Î≤îÏúÑÎ•º Î≤óÏñ¥ÎÇúÎ∂ÄÎ∂Ñ >>Ïä§Ìã∞Ïª§ Ï†úÍ±∞

```python
# -yÌÅ¨Í∏∞ÎßåÌÅº Ïä§Ìã∞Ïª§ crop
img_sticker = img_sticker[-refined_y:]
print (img_sticker.shape)
```

```python
# ÏõêÎ≥∏Ïù¥ÎØ∏ÏßÄÏùò Í≤ΩÍ≥ÑÍ∞íÏúºÎ°ú ÏàòÏ†ï
refined_y = 0
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

* sticker_areaÎäî ÏõêÎ≥∏Ïù¥ÎØ∏ÏßÄÏóêÏÑú Ïä§Ìã∞Ïª§Î•º Ï†ÅÏö©Ìï† ÏúÑÏπòÎ•º cropÌïú Ïù¥ÎØ∏ÏßÄ ÏûÖÎãàÎã§. ÏòàÏ†úÏóêÏÑúÎäî (202,0) Î∂ÄÌÑ∞ (202+130, 0+81) Î≤îÏúÑÏùò Ïù¥ÎØ∏ÏßÄÎ•º ÏùòÎØ∏Ìï©ÎãàÎã§.

Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÏÇ¨Ïö©Ìï† Î∂ÄÎ∂ÑÏùÄ 0 Ïù¥ ÏïÑÎãå ÏÉâÏù¥ ÏûàÎäî Î∂ÄÎ∂ÑÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§. Îî∞ÎùºÏÑú np.whereÎ•º ÌÜµÌï¥ img_sticker Í∞Ä 0 Ïù∏ Î∂ÄÎ∂ÑÏùÄ sticker_areaÎ•º ÏÇ¨Ïö©ÌïòÍ≥† 0Ïù¥ ÏïÑÎãå Î∂ÄÎ∂ÑÏùÑ img_stickerÎ•º ÏÇ¨Ïö©ÌïòÏãúÎ©¥ Îê©ÎãàÎã§. img_show Ïóê Îã§Ïãú Ï†ÅÏö©ÌïòÍ≤†ÏäµÎãàÎã§.

```python
# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóê Ïä§Ìã∞Ïª§ Ï†ÅÏö©
sticker_area = img_show[refined_y:img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_show[refined_y:img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
```

```python
# np.where()

# where(condition, [x, y])
# Return elements chosen from `x` or `y` depending on `condition`.

```
